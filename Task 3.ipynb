```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("householdtask3.csv")
```

```python
df.head(10)
```

```python
plt.scatter(df['year'], df['own'],color = 'purple')
plt.title("Scatter Plot")
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
plt.plot(df['year'])
plt.plot(df['own'])
plt.title("Line Chart")
plt.xlabel('year')
plt.ylabel('own')
plt.legend(['year','own'])
plt.show()
```

```python
plt.bar(df['year'], df['own'])
plt.title("Bar Chart")
plt.xlabel('year')
plt.ylabel('own')
plt.show()
```

```python
plt.hist(df['income'])
plt.title("Histogram")
plt.show()
```
